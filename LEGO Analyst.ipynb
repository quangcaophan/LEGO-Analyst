### Check data if it has any null values

```python
sets = pd.read_csv('C:\\Users\\Pearls\\Desktop\\Analyzing or something\\LEGO\\sets.csv')
themes = pd.read_csv('C:\\Users\\Pearls\\Desktop\\Analyzing or something\\LEGO\\themes.csv')
print(themes.info())
sets.info()
```

```python
for col in sets.columns:
    pct_missing = np.mean(sets[col].isnull())
    print('{} - {} %'.format(col, round(pct_missing*100)))
```

```python
for col in themes.columns:
    pct_missing = np.mean(themes[col].isnull())
    print('{} - {} %'.format(col, round(pct_missing*100)))
```

### Question 1 - What is the total number of parts per theme?

##### Merge data

```python
sets_themes = pd.merge(sets,themes, left_on = 'theme_id',right_on='id', how = 'inner',suffixes=('_sets','_themes'))
sets_themes.head()
```

```python
num_Parts_per_Theme = sets_themes.groupby('name_themes').sum().reset_index()
num_Parts_per_Theme = num_Parts_per_Theme[['name_themes', 'num_parts']].sort_values('num_parts', ascending = False)
num_Parts_per_Theme
```

### Question 2 - What is the total number of parts per year?

```python
num_parts_per_year = sets_themes.groupby(['year']).sum().reset_index()
num_parts_per_year = num_parts_per_year[['year','num_parts']].sort_values('num_parts',ascending=False)
num_parts_per_year
```

### Question 3 - How many sets were created in each Century in the dataset?

```python
unique_year = sets_themes['year'].unique()
unique_year.sort()
unique_year
```

```python
sets_themes['century'] = np.where(sets_themes['year'] > 2000,'21th century','20th century')
sets_themes.head(10)
```

```python
sets_created_each_century = sets_themes.groupby('century').count().reset_index()
sets_created_each_century[['century','name_sets']]
```

### Question 6 - What is the popular theme by year in terms of sets released in the 21st Century?

```python
themes_by_year = sets_themes.groupby(['year','name_themes']).count().reset_index()
themes_by_year = themes_by_year[['year','name_themes','set_num']].sort_values('set_num',ascending=False)
themes_by_year.drop_duplicates('year').sort_values('year', ascending = False)
```

### Question 7 - What is the most produced color of lego ever in terms of quantity of parts?

```python
colors = pd.read_csv('C:\\Users\\Pearls\\Desktop\\Analyzing or something\\LEGO\\colors.csv')
colors
```

```python
inventory = pd.read_csv('C:\\Users\\Pearls\\Desktop\\Analyzing or something\\LEGO\\inventory_parts.csv')
inventory
```

```python
inventory_color = inventory.merge(colors,right_on = 'id', left_on = 'color_id',how = 'left')
inventory_color
```

```python
popular_color = inventory_color.groupby(['name'])['quantity'].sum().reset_index()
popular_color.sort_values('quantity',ascending=False)
```
